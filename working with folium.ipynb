```python
#osm = open street map
map_osm = folium.Map(location = [-33.925,18.625], zoom_start=10)
map_osm
```

```python
map_dark=folium.Map(location=[-33.925,18.625], tiles = 'cartoDB dark_matter', zoom_start=10)
map_dark
```

```python
df_suburbs_locations = df_suburbs[["Latitude","Longitude"]]
```

```python
df_suburbs_locations.head()
```

```python
suburbs_locations_list = df_suburbs_locations.values.tolist()
```

```python
suburbs_locations_list
```

```python
suburbs_locations_list_size = len(suburbs_locations_list)
suburbs_locations_list_size
```

```python
for point in range(0,suburbs_locations_list_size):
    folium.Marker(suburbs_locations_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color="blue", icon_color="white", icon='tint')).add_to(map_osm)
map_osm
```

```python
#classification
for point in range(0,suburbs_locations_list_size):
    if df_suburbs['Oct 2017\nkl/month'][point] <5:
        folium.Marker(suburbs_locations_list[point], 
        popup = df_suburbs["Suburb"][point], 
        icon=folium.Icon(color='green', icon_color='white',
        icon='thumbs-up', angle=0, prefix='fa')).add_to(map_osm)
    elif df_suburbs['Oct 2017\nkl/month'][point] <5 and df_suburbs['Oct 2017\nkl/month'][point] >=8: 
        folium.Marker(suburbs_locations_list[point], 
        popup = df_suburbs["Suburb"][point], 
        icon=folium.Icon(color='blue', icon_color='white',
        icon='thumbs-up', angle=0, prefix='fa')).add_to(map_osm)
    else: 
        folium.Marker(suburbs_locations_list[point], 
        popup = df_suburbs["Suburb"][point], 
        icon=folium.Icon(color='red', icon_color='white',
        icon='thumbs-down', angle=0, prefix='fa')).add_to(map_osm)
```

```python
map_osm
```

```python
map_points = folium.Map(location= [-33.925, 18.625], zoom_start=10)

for point in range(0, suburbs_locations_list_size):
    folium.Circle(
    location = suburbs_locations_list[point], 
    popup = df_suburbs["Suburb"][point]+':' + str(df_suburbs["Oct 2017\nkl/month"][point]) + 'kl', 
    radius = str(df_suburbs["Oct 2017\nkl/month"][point]*100), 
    color='#17cbef',
    fill = True ,
    opacity=0.8 ,
    fill_color='#17cbef' ,
    stroke=True, 
    weight = 1.0).add_to(map_points)
```
