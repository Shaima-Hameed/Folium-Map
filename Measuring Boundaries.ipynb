```python
import pandas as pd
import numpy as np 
import folium 
```

```python
df_damlevels=pd.read_csv('https://raw.githubusercontent.com/EBISYS/Mapping-Dam-Levels/master/DamLevels.txt')
```

```python
df_damlevels.head()
```

```python
df_damlocation = pd.read_csv("https://raw.githubusercontent.com/EBISYS/Mapping-Dam-Levels/master/damLocations.txt")
```

```python
df_damlocation.head()
```

```python
df_damdata = pd.merge(df_damlevels, df_damlocation,on='Dam', how='left')
df_damdata
```

```python
df_dams = df_damdata[["Latitude","Longitude"]]
dams_list = df_dams.values.tolist()
dams_list_size = len(dams_list)
```

```python
dams_list
```

```python
dams_list_size
```

```python
map_dams = folium.Map(location=[-33.952,18.625] , zoom_start=10)
for point in range(0, dams_list_size):
    folium.Marker(dams_list[point], popup=df_damdata["Dam"][point]).add_to(map_dams)

map_dams
```

```python
map_url = folium.Map(location=[-34.078100,19.289200],
                    tiles = "http://services.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}",
                    zoom_start=10,
                    attr='Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community')
for point in range(0, dams_list_size):
    folium.Marker(dams_list[point], popup=df_damdata["Dam"][point]).add_to(map_url)
map_url
```

```python
from folium.plugins import MeasureControl 

map_url.add_child(MeasureControl())
```
